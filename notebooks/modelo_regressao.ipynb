Importa as bibliotecas necessárias e carrega o dataset já preparado, permitindo visualizar sua estrutura inicial.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

df = pd.read_csv("..\data\\processed\\spotify_prepared_v2_20251114_104521.csv")

df.head(20)
```

Remove as colunas textuais e mantém apenas as variáveis numéricas, definindo track_popularity como alvo e o restante como features.

```python
df_model = df.drop([
    "track_name",
    "playlist_genre",
    "is_hit",
    "is_hit_rule",
    "is_hit_agree"
], axis=1)

y = df_model["track_popularity"]

X = df_model.drop("track_popularity", axis=1)

print("X shape:", X.shape)
print("y shape:", y.shape)

X.head()
```

Separa a base em treino e teste para garantir uma avaliação correta do modelo com dados não vistos.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=27
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

Cria o modelo de regressão linear e realiza o treinamento usando o conjunto de treino.

```python
modelo = LinearRegression()

modelo.fit(X_train, y_train)

print("Modelo treinado com sucesso!")
```

Gera as previsões com os dados de teste e calcula as métricas R², MSE e RMSE para medir o desempenho do modelo.

```python
y_pred = modelo.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("R²:", r2)
print("MSE:", mse)
print("RMSE:", rmse)
```

Seleciona uma linha do conjunto de teste e compara a popularidade real da música com o valor previsto pelo modelo.

```python
i = 27

amostra = X_test.iloc[i]

predito = modelo.predict([amostra])[0]

print("POPULARIDADE REAL:", y_test.iloc[i])
print("POPULARIDADE PREVISTA:", predito)
```

Removendo algumas colunas e executando o teste novamente para aumentar o R² do algoritmo de regressão.

```python
limite = 0.01
df_num = df.select_dtypes(include=['number'])

corr = df_num.corr()['track_popularity'].abs()
colunas_fortes = corr[corr > limite].index.tolist()

df_filtrado = df_num[colunas_fortes]

print("Total de colunas antes:", df_num.shape[1])
print("Total de colunas depois:", df_filtrado.shape[1])

X = df_filtrado.drop("track_popularity", axis=1)
y = df_filtrado["track_popularity"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=27
)

modelo = LinearRegression()
modelo.fit(X_train, y_train)

y_pred = modelo.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("\n===== RESULTADOS APÓS FILTRAGEM =====")
print("R²:", r2)
print("MSE:", mse)
print("RMSE:", rmse)

i = 10
amostra = X_test.iloc[i]
pred = modelo.predict([amostra])[0]

print("\n===== TESTE COM AMOSTRA =====")
print("POPULARIDADE REAL:", y_test.iloc[i])
print("POPULARIDADE PREVISTA:", pred)
```

# Análise Final — Projeto Spotify (Regressão Linear)

Durante o desenvolvimento do projeto, aplicamos o algoritmo de Regressão Linear com o objetivo de avaliar se seria possível identificar se uma música é “hit” ou “não hit” utilizando seus atributos numéricos e a variável track_popularity.

## Função da Regressão Linear

A regressão linear é um algoritmo de aprendizado supervisionado voltado para prever valores contínuos. No contexto deste projeto, ela foi utilizada para estimar o nível de popularidade das músicas, representado pela coluna track_popularity.

Assim, a regressão responde a perguntas como:

“Qual será aproximadamente a popularidade desta faixa?”

“Essa música tende a apresentar popularidade mais alta ou mais baixa?”

## Limitações do Algoritmo para o Problema Proposto

Apesar de fornecer previsões numéricas, a regressão linear não é um método de classificação. Portanto, ela não é capaz de determinar diretamente se uma música se enquadra na categoria “hit” ou “não hit”.

Para isso, seria necessário transformar o problema em uma classificação binária, definindo um critério, por exemplo:

hit = 1, quando track_popularity ≥ determinado limite;

não hit = 0, caso contrário.

Como trabalhamos exclusivamente com regressão, o modelo aprendeu apenas a prever valores contínuos e não a categorizar músicas em grupos distintos.

## Conclusão

A Regressão Linear foi eficaz para estimar o nível de popularidade das músicas. No entanto, o algoritmo não permite identificar automaticamente se uma música é hit ou não, pois esse tipo de resposta exige um modelo de classificação e não de regressão. A Regressão Linear é adequada para prever valores de popularidade, mas não atende ao objetivo de classificar músicas como hit ou não hit. Para isso, é indispensável o uso de um modelo de classificação.
