```python
import pandas as pd

file_high = pd.read_csv("/content/high_popularity_spotify_data.csv")
file_low = pd.read_csv("/content/low_popularity_spotify_data.csv")
```

```python
file_high['Hit'] = 'sim'
file_low ['Hit'] = 'nao'
```

```python
final_file = pd.concat([file_high, file_low], ignore_index=True, sort=True)
colunas_remover = [
    'id',
    'playlist_id',
    'playlist_name',
    'type',
    'track_id',
    'uri',
    'analysis_url',
    'track_href',
    'track_popularity',
    'track_album_id',
    'track_album_name',
    'track_album_release_date',
    'track_artist',
    'track_name'
]
final_file.drop(columns=colunas_remover, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le_genre = LabelEncoder()
le_sub = LabelEncoder()

final_file['playlist_genre'] = le_genre.fit_transform(final_file['playlist_genre'])
final_file['playlist_subgenre'] = le_sub.fit_transform(final_file['playlist_subgenre'])
```

```python
x = final_file.drop(columns="Hit")
y = final_file["Hit"]
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, Y_treino, Y_teste = train_test_split(x,y, stratify=y, train_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

modelo = DecisionTreeClassifier()
modelo.fit(X_treino,Y_treino)
predicao = modelo.predict(X_teste)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_teste, predicao)
accuracy
```

```python
resultado_final = X_teste.copy()
resultado_final['Real'] = Y_teste.values
resultado_final['Predicao'] = predicao
resultado_final
```
